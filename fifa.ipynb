```python
%matplotlib inline
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
players = pd.read_csv('player_stats.csv', encoding = 'latin-1')
```

```python
players.info()
```

```python
players.head()
```

```python
players['count'] = 1
```

```python
by_country = players.groupby(['country'])[['count']].sum().sort_values(by = "count", 
                                                                            ascending=False)
by_country.head(10)
```

```python
fig, ax = plt.subplots(figsize = (6, 3))
ax.bar(by_country.head(10).index, by_country['count'].head(10), color = 'red')
ax.set(title = "Top 10 paises con más jugadores en el fifa 24")
plt.xticks(rotation = 75)
```

```python

by_height = players.groupby(['player'])[['height']].max().sort_values(by = 'height',
                                                                        ascending = False)
by_height.head(10)
```

```python
fig, ax = plt.subplots(figsize = (6, 3))
ax.bar(by_height.head(5).index, by_height['height'].head(5), color = 'red')
ax.set(title = "Top 10 jugadores  más altos en el fifa 24")
plt.xticks(rotation = 75)
```

```python
by_height_country = players.groupby(['country'])[['height']].max().sort_values(by = 'height',
                                                                        ascending = False)
by_height_country.head(10)
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax.pie(by_height_country['height'].head(5), labels= by_height_country.head(5).index, 
       autopct='%1.1f%%', startangle=180, shadow = True);
ax.set(title="Porcentaje de paises con jugadores más altos");
```

```python

by_height_club = players.groupby(['club'])[['height']].max().sort_values(by = 'height',
                                                                        ascending = False)
by_height_club.head(10)
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax.pie(by_height_club['height'].head(10), labels= by_height_club.head(10).index, 
       autopct='%1.1f%%', startangle=180, shadow = True);
ax.set(title="Porcentaje de clubes con jugadores más altos");
```

```python
by_weight = players.groupby(['player'])[['weight']].max().sort_values(by = 'weight',
                                                                        ascending = False)
by_weight.head(10)
```

```python
fig, ax = plt.subplots(figsize = (6, 3))
ax.bar(by_weight.head(10).index, by_weight['weight'].head(10), color = 'red')
ax.set(title = "Top 10 jugadores  más pesados en el fifa 24")
plt.xticks(rotation = 75)
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax.pie(by_weight_country['weight'].head(5), labels= by_weight_country.head(5).index, 
       autopct='%1.1f%%', startangle=180, shadow = True);
ax.set(title="Porcentaje de paises con jugadores más pesados");
```
